# import libraries

```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

# read the dataset

```python
df = pd.read_csv('Iris.csv')
df
```

```python
x  = df.drop(['Id','Species'],axis=1)
x
```

# scaling

```python
std_scaler = StandardScaler()
array = std_scaler.fit_transform(x)
array
x_df = pd.DataFrame(array,columns=x.columns)
```

```python
x_df
```

# Model training

```python
kmeans_model = KMeans(n_clusters=3)
kmeans_model.fit(x_df)
```

```python
y_var = kmeans_model.fit_predict(x_df)
y_var
```

```python
class1 = np.where(y_var==1)
class1
```

```python
x.iloc[class1]
```

```python
class2 = np.where(y_var==2)
class2
```

```python
x.iloc[class2]
```

```python
WCCS = kmeans_model.inertia_  # k = 3
WCCS
```

```python
wcss_list = []
for k in range(1,10):
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(x_df)
    wcss = kmeans_model.inertia_
    wcss_list.append(wcss)
    print('K =',k,  'WCSS = ',wcss)
    
```

```python
wcss_list
```

```python
k = range(1,10)
plt.plot(k,wcss_list)
plt.xlabel('Number of cluster(k)')
plt.ylabel('WCSS')
plt.title('Elbow method graph')
```
